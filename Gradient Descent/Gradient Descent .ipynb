```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the cost function
def cost_function(theta):
    return theta**2

# Gradient descent function
def gradient_descent(theta, learning_rate, num_iterations):
    for _ in range(num_iterations):
        gradient = 2 * theta
        theta -= learning_rate * gradient
    return theta

# Generate data for visualization
theta_values = np.linspace(-5, 5, 100)
cost_values = np.array([cost_function(theta) for theta in theta_values])

# Visualize the cost function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(theta_values, theta_values)
Z = np.array([[cost_function(theta) for theta in row] for row in X])
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)
ax.set_xlabel('theta')
ax.set_ylabel('theta')
ax.set_zlabel('Cost')

# Perform gradient descent and visualize the path
initial_theta = 3
learning_rate = 0.1
num_iterations = 100
final_theta = gradient_descent(initial_theta, learning_rate, num_iterations)

theta_path = np.linspace(initial_theta, final_theta, num_iterations)
cost_path = np.array([cost_function(theta) for theta in theta_path])

ax.plot(theta_path, theta_path, cost_path, color='red', linewidth=2)
plt.title('Gradient Descent Visualization')
plt.show()
```
